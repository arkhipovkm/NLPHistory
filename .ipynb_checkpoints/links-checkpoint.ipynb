```python
import pandas as pd
```

```python
df_indexed = pd.read_pickle('dcdf.pickle')
```

```python
df = df_indexed.reset_index()
df.head()
```

```python
df['post'] = df.club_id.astype(str) + '_' + df.post_id
```

```python
df['post_link'] = 'https://vk.com/public' + df.club_id.astype(str) + '?w=wall-' + df.post
```

```python
wow = df[df.rubric_id==36]
w = wow.loc[:,['date', 'text', 'post_link']]
w[w.date > '2018-01-01'].sort_values('date').shape
```

```python
df.groupby(['post_link']).count().sort_values(by='comment_id', ascending=False).comment_id.to_excel('posturl_counts.xls')
```
